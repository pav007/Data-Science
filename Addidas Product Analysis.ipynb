```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import plotly.express as px
color=sns.color_palette('bright')
plt.pie(df["Retailer"].value_counts().values,labels=df["Retailer"].value_counts().index,colors=color,autopct='%.0f%%')
plt.show()
```

```python
#From the above pie diagram we could able to infere that most of the sales were done by West Gear and Least by Southeast
```

```python
#Pie chart for better undertsading of the sales made in a region
color=sns.color_palette("bright")
plt.pie(df['Region'].value_counts().values,labels=df['Region'].value_counts().index,colors=color,autopct='%.0f%%')
plt.show()
```

```python
#From the above pie diagram we could able to infere that most of the sales were done by Northeast and West Regions
#Least were done by Southeast
```

```python
x=['State','City','Product']
for y in x:
    dig = df[[y,'Operating Margin']].groupby(y).sum()
    dig.plot(kind='bar',figsize=(10,7),title='Total Sales performed by '+y, colormap='rainbow')
```

```python
x=['State','City']
for i in x:
    dig=df[[i,'Operating Profit']].groupby(i).sum()
    dig.plot(kind='bar',figsize=(10,7),title='Total Sales performed by '+i, legend=False,colormap='jet')
```

```python
#units sold analysis
x=['State','City']
for i in x:
    dig=df[[i,'Units Sold']].groupby(i).sum()
    dig.plot(kind='bar',figsize=(10,7),title='Total Sales performed by '+i, legend=False,colormap='pink')
```
